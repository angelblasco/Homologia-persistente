## 1. Motivación

 - Sea $(X, d)$ un **espacio métrico** (desconocido). Típicamente $X\subseteq \mathbb R^d$.

 - Sea $X_n\subseteq X$ una **muestra aleatoria** de $n$ puntos. Asumimos conocida $d|_{X_n\times X_n}$.

 - Queremos inferir la **homología** del espacio $X$ a partir de la muestra $X_n$.

![SegmentLocal](images/point_cloud.gif "segment")

_(Gifts extraidos de https://www.youtube.com/watch?v=h0bnG1Wavag)_

 **Problema 1:** $X_n$ como espacio topológico tiene homología $\mathbb Z^n$ en grado cero y trivial en otros grados.

 **Idea 1:**
 - 'Estimar' a $X$ como $U_{\varepsilon} = \displaystyle \bigcup_{i=1}^n B(x_i, \varepsilon)$ para algún $\varepsilon$.
 
 ![SegmentLocal](images/union.gif "segment")
 
- Calcular la homologia de $\displaystyle \bigcup_{i=1}^n B(x_i, \varepsilon)$ como la homología del nervio del cubrimiento $\mathcal{U}_{\varepsilon} = \{B(x_i, \varepsilon)\}$

![SegmentLocal](images/nervio.gif "segment")

 **Problema 2:** ¿Cómo elegir $\varepsilon$?

![SegmentLocal](images/epsilon.gif "segment")

**Teorema (Niyogi, Smale & Weinberger, 2008):** Sea $\mathcal M$ una subvariedad compacta de dimensión $k$ de $\mathbb R^d$ con $\mathrm{rch}(\mathcal M) = \tau$. Sea $X_n = \{x_1, x_2, \dots x_n\}$ una muestra aleatoria i.i.d. respecto de una densidad uniforme en $\mathcal M$. 

Sea $\varepsilon <\tau/2$. Sea  
$$U_{\varepsilon} = \displaystyle \bigcup_{i=1}^n B(x_i, \varepsilon).$$ 
Dado $\delta >0$, si 

$$n>\beta_1\left(\log(\beta_2) + \log(1/\delta)\right)$$ 

entonces con probabilidad mayor que $1-\delta$, 

$$U_{\varepsilon}\simeq \mathcal M,$$
donde $$\beta_1 = \frac{\mathrm {vol} \mathcal M}{\cos^k(\theta_1)\mathrm{vol}(B^k_{\varepsilon/4})}, ~~~\beta_1 = \frac{\mathrm {vol} \mathcal M}{\cos^k(\theta_2)\mathrm{vol}(B^k_{\varepsilon/8})},$$

$$\theta_1=\arcsin(\varepsilon/8\tau), ~~ \theta_2=\arcsin(\varepsilon/16\tau)$$

 **Idea 2:** En lugar **de elegir $\varepsilon$**, calculamos la homología de $U_{\varepsilon}$ **para todos** los $\varepsilon$ 'de manera coherente'.

## 2. Filtración de espacios

Una **filtración de complejos simpliciales** es una sucesión $\{C_i\}_{i\in I}$ de complejos simpliciales junto con inclusiones $\iota_i:C_i\to C_{i+1}$.

Decimos que $X$ es un **complejo simplicial filtrado** si existe una filtración $\{C_i\}_{i\in I}$ de manera que para $i\geq i_0$, $C_i = X$.

<img src="images/filtration.png" width="800">

#### 2.1 Ejemplos de filtraciones a partir de $X_n$:

- **Filtración de Cech:** Para cada $\varepsilon>0$, sea $\mathcal{U}_{\varepsilon} = \{B(x_i, \varepsilon)\}$ una colección de abiertos. Consideramos

$$\check C_{\varepsilon} = \mathcal{N}(\mathcal{U}_{\varepsilon})$$


- **Filtración de Vietoris-Rips:** Consideramos $\Delta_n$ el $(n-1)$-simplex con vértices en $X_n$. Definimos

$$V_{\varepsilon} = \{\sigma  \in \Delta_n: \mathrm{diam}(\sigma)<\varepsilon\}$$


**Propiedad:** $$\check C_{\varepsilon}\subseteq V_{\varepsilon}\subseteq \check C_{2\varepsilon}$$


<img src="images/cech_rips.png" width="600">

## 3. Homología persistente

Sea $k$ un cuerpo. Sea $C_i$ una filtración de $X$.

Para cada $i$, la inclusión $\iota_i:C_i\to C_{i+1}$ induce un morfismo en las homologías:
        $$\iota_{*}:H_{*}(C_i)\to H_{*}(C_{i+1})$$

Para cada grado $j$, se define el módulo de persistencia
$$M_j = \bigoplus_{i\in I} H_j(C_i)$$

$M_j$ se puede ver como un $k[x]$-módulo graduado, donde $x$ actúa como _shift map_.

Por **Teorema de Estructura** para módulos sobre un DIP graduado
$$M_j = 
\left( \bigoplus_{s=1}^n \Sigma^{\alpha_s} k[x]\right) \oplus \left(\bigoplus_{l=1}^m \Sigma^{\gamma_l}k[x]/x^{n_l}k[x]\right)$$

donde $\alpha_s, \gamma_j\in \mathbb{Z}$, $n_l<n_{l+1}$ y  $\Sigma^{\alpha}$ denota un _$\alpha$-shift_ en el grado.

Representamos $M_j$ mediante un conjunto de intervalos $(\alpha_s, + \infty)$ y $(\gamma_l,\gamma_l+n_l)$, que denominamos **barcode**.



El **cálculo** se realiza usando álgebra lineal. La complejidad es $O(N^3)$, con $N$ la cantidad de símplices de $X$.

### Círculos

```python
from sklearn import datasets
circles, labels = datasets.make_circles(n_samples=1000, noise=0.06, factor=0.5)
print('Data dimension:{}'.format(circles.shape))
```

```python
import matplotlib.pyplot as plt; 
import seaborn as sns
sns.set()

fig = plt.figure()
ax = fig.add_subplot(111)
ax = sns.scatterplot(x=circles[:,0], y=circles[:,1],   s=15)
plt.title('Two circles dataset \n N=%s'%(circles.shape[0]))
plt.show()
```

Calculamos la filtración de Vietoris-Rips

```python
%%time
# The RipsComplex() function creates a one skeleton graph from the point cloud.
Rips_complex = gd.RipsComplex(circles, max_edge_length=0.6) 
```

```python
%%time
# The create_simplex_tree() method creates the filtered complex.
Rips_simplex_tree = Rips_complex.create_simplex_tree(max_dimension=2) 
```

```python
%%time
# The get_filtration() method computes the simplices of the filtration
filt_Rips = list(Rips_simplex_tree.get_filtration())
```

Veamos el complejo de Vietoris-Rips para un valor de $\varepsilon$

```python
import numpy as np

eps = 0.15
triangles = np.array([s[0] for s in Rips_simplex_tree.get_skeleton(2) if len(s[0])==3 and s[1] <= eps])
```

```python
plt.triplot(circles[:,0], circles[:,1], triangles)
plt.plot(circles[:,0], circles[:,1], '.')
plt.show()
```

Calculamos el diagrama de persistencia

```python
%%time
# We can compute persistence on the simplex tree structure using the persistence() method
diag_Rips = Rips_simplex_tree.persistence()
```

```python
%%time
gd.plot_persistence_diagram(diag_Rips)
```

## 5. Distancia Bottleneck

Dados dos diagramas de persistencia $\mathrm{dgm_1}$ y $\mathrm{dgm_2}$ para cierto grado, se define la **distancia bottleneck** entre los diagramas como:

$$d_b(\mathrm{dgm_1},\mathrm{dgm_2}) = \inf_{M \text{ matching }} \sup_{(x,y)\in M} ||x-y||_{\infty}$$

donde $M\subseteq \mathrm{dgm_1}\times \mathrm{dgm_2}$ es un matching, y se considera que los puntos de la diagonal pertenecen a ambos diagramas.

```python
import persim
distance_bottleneck, (matching, D) = persim.bottleneck(dgm_clean, dgm_noisy, matching=True)
persim.bottleneck(dgm_clean, dgm_noisy)
```

```python
sns.set()
persim.bottleneck_matching(dgm_clean, dgm_noisy, matching, D, labels=['Clean $H_1$', 'Noisy $H_1$'])
```

Veamos cómo evoluciona la distancia Bottleneck entre los diagramas de las muestras con y sin ruido paradistinos niveles de ruido.

```python
import ripser

fig = plt.figure(figsize = (10,10))

for i, n in enumerate([0.01, 0.05, 0.1, 0.15, 0.2, 0.3, 0.4, 0.6, 0.8]):
    sns.set()
    plt.subplot(331+i)

    ds_clean = tadasets.dsphere(d=1, n=100, noise=0.0)
    dgm_clean = ripser.ripser(ds_clean)['dgms'][1]

    ds = tadasets.dsphere(d=1, n=100, noise=n)
    dgm = ripser.ripser(ds)['dgms'][1]
    d, (matching, D) = persim.bottleneck(
        dgm_clean,
        dgm,
        matching=True
    )
    persim.bottleneck_matching(dgm_clean, dgm, matching, D, labels=['Clean $H_1$', 'Noisy $H_1$'])

    plt.title("Noise:{} Distance:{:.3f}".format(n, d))

plt.tight_layout()
```

## 6. Estabilidad

![SegmentLocal](images/stability.gif "segment")

Recordar que la **distancia de Gromov-Hausdorff** entre dos espacios métricos compactos $(X, d_X)$ e $(Y, d_Y)$ es:

$$d_{GH}(X,Y) = \frac{1}{2}\inf_{R\subseteq X\times Y}\sup_{(x,y), (x', y')\in R}|d_X(x,x')-d_Y(y,y')|$$

donde $R$ es una _correspondencia_ (ie, $\forall x\in X, \exists y\in Y$ tal que $(x,y)\in R$ y el análogo para $Y$)
 
Equivalentemente, 

$$d_{GH}(X,Y) = \inf_{(Z, d_Z)}\{d_H\left(\phi_X(X),\phi_Y(Y)\right):\phi_X:X\to Z,\phi_Y:Y\to Z \text{ isometrias}\}$$

donde $d_H$ denota la distancia de Hausdorff.


- Frédéric Chazal, Vin de Silva, Marc Glisse, and Steve Oudot. (2012) **The structure and stability of persistence modules**.

**Teorema (Chazal et.al, 2012)**: Si $X, Y$ son dos espacios métricos compactos, entonces 

$$d_b(\mathrm{dgm(Filt}(X)), \mathrm{dgm(Filt}(Y)))\leq 2 d_{GH}(X,Y)$$

- Fasy, B.T., Lecci, F., Rinaldo, A., Wasserman, L., Balakrishnan, S., & Singh, A. (2014). **Confidence sets for persistence diagrams.** Annals of Statistics, 42, 2301-2339.

**Teorema (Fasy et.al 2014)**: Dado $\alpha>0$, existe $c_n = c(X_n)$ tal que 


$$\liminf_{n\to\infty} P\big(d_b\mathrm{(dgm(Filt}(X_n), \mathrm{dgm(Filt}(X))\in [0, c_n]\big)\geq 1-\alpha$$

Es decir, $[0,c_n]$ es un intervalo de confianza asintótico para $d_b\mathrm{(dgm(Filt}(X_n), \mathrm{dgm(Filt}(X))$.

**Teorema (Chazal et.al. 2015):**

Sea $(M, \rho)$ un espacio métrico. Sea $X_n$ una muestra i.i.d. de $M$ de acuerdo a una medida de probabilidad $\mu$ desconocida. Sea $X_{\mu}\subseteq M$ el soporte de $\mu$. Supongamos que $\mu$ cumple:
- $X_{\mu}$ es compacto;
- existen $a,b>0$ tales que para todo $r>0$, $x\in X_{\mu}$, $\mu(B(x,r))>\min(ar^b, 1).$

Entonces,

$$E[d_b(\mathrm{dgm(Filt}(X_{\mu})), \mathrm{dgm(Filt}(X_n))]\leq C\left(\dfrac{\ln(n)}{n}\right)^{1/b}$$

donde $C$ solo depende de $a$ y de $b$ (no de $M$).
