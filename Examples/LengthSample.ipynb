# Subsampling

This notebook shows how the lenght of the sample affects the computation of Persistent Homology

```python
import numpy as np
import matplotlib.pyplot as plt
import tadasets
```

```python
from ripser import ripser
from persim import plot_diagrams
```

```python
l = [50, 100, 300, 500]
for i,n in enumerate(l):
    sphere = tadasets.sphere(n, r = 10)
    plt.subplot(221+i)
    dgm = ripser(sphere, maxdim = 2)['dgms']
    plot_diagrams(dgm)
```
