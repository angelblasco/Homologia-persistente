# Circles

This notebook introduces the basic concepts in order to see how Persistent Homology captures the topology of a point cloud.

We use a toy data set (two circles) in order to illustrate. 

 **Contents**:
        
    - Data generation
    - Persistence diagrams

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

## Data generation

Generate the Circles data set.

```python
data, labels = datasets.make_circles(n_samples=3000, noise=0.06, factor=0.5)
print('Data dimension:{}'.format(data.shape))
```


Visualize the data

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax.scatter(x=data[:,0], y=data[:,1], c=labels, s=2)
plt.title('Two circles dataset \n N=%s'%(data.shape[0]))

plt.show()
```

## Persistent Homology

```python
#!pip install ripser
from ripser import Rips
```

```python
%%time
rips = Rips()
dgms = rips.fit_transform(data)
rips.plot(dgms)
```
